# **Hypothesis** **Testing**

Question 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing?

        
    -->What is a null hypothesis (H₀) and why is it important?

The null hypothesis (H₀) is a statement that there is no effect, no difference, or no change in the population.
It serves as the starting assumption in hypothesis testing.

Importance:
	•	It provides a basis for statistical testing.
	•	Helps determine whether observed results are due to chance or a real effect.
	•	Decisions are made by either rejecting or failing to reject H₀.


Question 2: What does the significance level (α) represent in hypothesis testing?

    -->The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error).
	•	Common values: 0.05, 0.01
	•	α = 0.05 means there is a 5% risk of making a wrong rejection.

Question 4: Explain the difference between a one-tailed and two-tailed test. Give an example of each.


    -->Difference between one-tailed and two-tailed tests

One-tailed test
	•	Tests for an effect in one direction only
	•	Example:
H₀: μ = 50
H₁: μ > 50

Two-tailed test
	•	Tests for an effect in both directions
	•	Example:
H₀: μ = 50
H₁: μ ≠ 50


Question 5: A company claims that the average time to resolve a customer complaint is 10 minutes.

A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes.

At α = 0.05, test the claim.


    -->Testing the company’s claim

Given:
	•	Claimed mean (μ) = 10 minutes
	•	Sample mean (x̄) = 12
	•	Standard deviation (s) = 3
	•	Sample size (n) = 9
	•	α = 0.05

Step 1: Hypotheses
	•	H₀: μ = 10
	•	H₁: μ ≠ 10 (two-tailed)

Step 2: Test statistic (t-test)

t = \frac{12 - 10}{3/\sqrt{9}} = \frac{2}{1} = 2

Step 3: Critical value
	•	df = 8
	•	t₀.₀₅ = ±2.306

Step 4: Decision
	•	|2| < 2.306 → Fail to reject H₀

Conclusion

There is not enough evidence to reject the company’s claim at 5% significance level.


Question 6: When should you use a Z-test instead of a t-test?


    -->Use a Z-test when:
	•	Population standard deviation (σ) is known
	•	Sample size is large (n ≥ 30)
	•	Data is approximately normally distributed

Use a t-test when:
	•	σ is unknown
	•	Sample size is small (n < 30)
